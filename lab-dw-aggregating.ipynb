4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
Life_Time_Value = marketing_customer_analysis[['Education','Gender','Customer Lifetime Value']].pivot_table(index= ['Education','Gender'], aggfunc = ['max', 'min', 'median'])
Life_Time_Value
```

```python
# Most of the customer life time value data is organized around USD 5332 and USD 6286. There are big outliers in each category.
Life_Time_Value1 = marketing_customer_analysis[['Education','Gender','Customer Lifetime Value']].groupby(['Education','Gender']).agg({'max', 'min', 'median'})
Life_Time_Value1
```
